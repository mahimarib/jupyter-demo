```python
%pylab inline
import pandas
import seaborn
```

# Load CVS File in Memory

```python
data = pandas.read_csv(r'C:\Users\Home\Desktop\School Work\uber-raw-data-aug14.csv')
```

### Converted Date/Time and add some useful columns

```python
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

### Gets the Day of the Month

```python
def get_date_of_month(date_time):
    return date_time.day

data['Date of Month'] = data['Date/Time'].map(get_date_of_month)
```

### Gets the Weekday and Hour

```python
def get_weekday(date_time):
    return date_time.weekday()

data['Weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(date_time):
    return date_time.hour

data['Hour'] = data['Date/Time'].map(get_hour)
```

```python
dateTime = pandas.to_datetime(dateTime)
```

```python
def count_rows(rows):
    return len(rows)

by_date = data.groupby('Date of Month').apply(count_rows)
bar(range(1, 32), by_date)
```

```python
hist(data['Date of Month'], bins=31, rwidth=.8, range=(0.5,31.5))
xlabel('Days of the Month')
ylabel('Frequency')
title('Uber Aug 2014 London')
```

```python
by_cross = data.groupby(['Weekday', 'Hour']).apply(count_rows).unstack()
```

```python
seaborn.heatmap(by_cross)
```
